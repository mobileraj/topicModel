```python
import csv
data={}
foo=open('/home/raj/projects/topicModels/Topics.csv','w')
def getContrib(d):
    l=len(d)
    x=['0']*5
    i=2
    while i < l:
        x[int(d[i])-1]=d[i+1]
        i = i+2
    return x

with open('/home/raj/projects/topicModels/output_csv/TopicsInDocs.csv') as tmp:
    # Skip first line (if any)
    next(tmp, None)
    foo.write('year,topic1,topic2,topic3,topic4,topic5\n')
    # {line_num: row}
    for x in tmp:
        d=x.strip().split(',')
        yr = d[1][38:42]
        data.setdefault(yr,None)
        data[yr]=getContrib(d)
        foo.write(yr+','+','.join(getContrib(d))+'\n')
foo.close()
```
